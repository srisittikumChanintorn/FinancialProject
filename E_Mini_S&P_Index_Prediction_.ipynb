- ### **E-Mini S&P 500**
#### &nbsp;&nbsp;S&P 500 — ย่อมาจาก Standard and Poor's 500, เป็นดัชนีของตลาดหุ้นของสหรัฐอเมริกา ที่ขึ้นกับมูลค่าตลาดของบริษัท 500 อันดับแรกที่มีหุ้นสามัญจดทะเบียนอยู่ในตลาดหลักทรัพย์นิวยอร์ก หรือ NASDAQ ดัชนี S&P 500 ทำมาเพื่อสะท้อนสถานะของตลาดหุ้นทั้งหมด โดยการติดตามผลตอบแทนและความผันผวนของหุ้นบริษัทขนาดใหญ่ (large cap) 500 อันดับแรกบน NYSE ซึ่งคิดเป็นกว่า 80% ของมูลค่าตลาดทั้งหมด บริษัท 500 แห่ง ประกอบไปด้วย SP500 ทั่วทั้งอุตสาหกรรมหลักทั้งหมด และถูกเลือกตามมูลค่าตลาด สภาพคล่อง ความสามารถในการซื้อขาย และปัจจัยอื่นๆ. การติดตาม S&P 500 เป็นที่นิยมในการวัดและเปรียบเทียบประสิทธิภาพการลงทุนในตลาดหุ้น ซึ่งสถานะของ S&P 500 ถือเป็นตัวบ่งชี้ว่าตลาดหุ้นในสหรัฐอเมริกากำลังเติบโตหรือลดลง นอกจากนี้ยังเป็นตัวบ่งชี้ที่นำมาใช้ในการวิเคราะห์และบริหารการลงทุนในหุ้นและกองทุนรวม

- ### CRUDE OIL FUTURE
#### &nbsp;&nbsp;สัญญาซื้อขายน้ำมันดิบเบาล่วงหน้า (Light Sweet Crude Oil Futures Contract) เป็นสัญญาซื้อขายที่เกี่ยวข้องกับการซื้อขายน้ำมันดิบประเภทเบาโดยเฉพาะ ซึ่งในทางเทคนิคสัญญาซื้อขายน้ำมันดิบเบาล่วงหน้านี้สามารถใช้เป็นเครื่องมือในการซื้อขายและการลงทุนในตลาดน้ำมันดิบได้  ซึ่งสัญญาซื้อขายน้ำมันดิบเบาล่วงหน้าจะระบุราคาที่ผู้ซื้อจะชำระเมื่อสัญญาเสร็จสิ้นในอนาคต โดยส่วนใหญ่จะมีระยะเวลาที่กำหนดไว้ล่วงหน้า เช่น 1 เดือนหรือ 3 เดือน   การซื้อขายสัญญาซื้อขายน้ำมันดิบเบาล่วงหน้าเกิดขึ้นในตลาดการเงินเช่น ตลาดสินค้าอนุพันธ์ ซึ่งมีสถานที่การซื้อขายที่กำหนดไว้เช่น ตลาดสินค้าดิบแห่งนิวยอร์ก (New York Mercantile Exchange, NYMEX) และตลาดอุตสาหกรรมการซื้อขายสินค้า (Intercontinental Exchange, ICE) เป็นต้น

- ### GOLD FUTURE
#### &nbsp;&nbsp;Gold Futures (สัญญาซื้อขายทองคำล่วงหน้า) เป็นสัญญาทางการเงินที่เกี่ยวข้องกับการซื้อขายทองคำในอนาคต สัญญาซื้อขายทองคำล่วงหน้านี้ถูกใช้ในตลาดการเงินเพื่อการลงทุนหรือการป้องกันความเสี่ยงทางการเงินที่เกี่ยวข้องกับราคาทองคำ ตลาดทองคำล่วงหน้าสามารถซื้อขายในแพลตฟอร์มการซื้อขายออนไลน์ที่กำหนดไว้เช่น COMEX (Commodity Exchange, Inc.) ในสหรัฐอเมริกา หรือตลาดอื่นๆ ที่มีการซื้อขายสัญญาซื้อขายทองคำล่วงหน้า การซื้อขายสัญญาซื้อขายทองคำล่วงหน้าเป็นวิธีหนึ่งในการลงทุนในทองคำ ผู้ลงทุนสามารถใช้สัญญาซื้อขายทองคำล่วงหน้าเพื่อดำเนินการซื้อหรือขายทองคำโดยไม่ต้องมีการถือทองคำสัมมนา นอกจากนี้สัญญาซื้อขายทองคำล่วงหน้ายังสามารถใช้เป็นเครื่องมือในการป้องกันความเสี่ยงทางการเงินที่เกี่ยวข้องกับราคาทองคำในอนาคต

- ### CBOE 10 YR TREASURY NOTE YIELD
#### &nbsp;&nbsp;TNX เป็นรหัสสัญญาซื้อขายแลกเปลี่ยน (ticker symbol) สำหรับสัญญาซื้อขายพันธบัตรค่าเงินต้นต่อระยะเวลา 10 ปีของสหรัฐอเมริกา (10-Year Treasury Note) ซึ่งเป็นหนึ่งในสินทรัพย์ที่ใช้ในการประเมินอัตราดอกเบี้ยของสหรัฐอเมริกาในระยะยาว อัตราดอกเบี้ยและดัชนีหุ้นมีความสัมพันธ์กันในบางกรณี ซึ่งอัตราดอกเบี้ยสามารถมีผลต่อผลตอบแทนในการลงทุนในหลักทรัพย์หรือหุ้น นี่คือบางปัจจัยที่มีความสัมพันธ์ระหว่างอัตราดอกเบี้ยกับดัชนีหุ้น:
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **ผลตอบแทนที่ลงทุน: อัตราดอกเบี้ยมีผลต่อผลตอบแทนที่คาดหวังจากการลงทุนในหลักทรัพย์หรือหุ้น หากอัตราดอกเบี้ยสูงขึ้น อาจส่งผลให้ผลตอบแทนจากการลงทุนในหุ้นลดลง เนื่องจากการลงทุนในหุ้นมีความเสี่ยงสูงกว่าการลงทุนในตราสารหนี้ที่มีอัตราดอกเบี้ยสูง**
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **ความเคลื่อนไหวของตลาด: อัตราดอกเบี้ยสามารถส่งผลต่อความเคลื่อนไหวของตลาดหุ้น หากอัตราดอกเบี้ยเพิ่มขึ้นอาจส่งผลให้นักลงทุนกลับมาลงทุนในตราสารหนี้แทนการลงทุนในหุ้น ซึ่งอาจทำให้ดัชนีหุ้นลดลง**
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **สภาวะเศรษฐกิจ: อัตราดอกเบี้ยสามารถสะท้อนสภาวะเศรษฐกิจทั้งในระดับประเทศและระดับโลก หากเศรษฐกิจเจริญเติบโตและเข้มแข็ง อัตราดอกเบี้ยอาจมีแนวโน้มเพิ่มขึ้น และอาจส่งผลให้ดัชนีหุ้นเพิ่มขึ้นด้วยผลกระทบจากเหตุการณ์ทางเศรษฐกิจและการเมือง: เหตุการณ์ทางเศรษฐกิจและการเมือง เช่น นโยบายการเงินของธนาคารกลาง นโยบายการเงินระหว่างประเทศ ข่าวสารทางเศรษฐกิจ หรือเหตุการณ์ทางการเมืองสำคัญ อาจมีผลต่ออัตราดอกเบี้ยและส่งผลต่อดัชนีหุ้น**

- ### CBOE Volatility Index (VIX)
#### &nbsp;&nbsp;Volatility Index หรือ (VIX) คือดัชนีซึ่งคำนวณโดยตลาดซื้อขายอนุพันธ์ Chicago Board Options Exchange  ซึ่ง CBOE Volatility Index เป็นดัชนีที่วัดและติดตามความผันผวนในตลาดหุ้นในระดับสัมพันธ์ของตลาดทางการเงินในสหรัฐอเมริกา รูปแบบการคำนวณของ VIX มีความเชื่อมโยงกับราคาตัวเลือกซื้อขายตัวเลือกในตลาดหุ้น ทำให้เป็นตัวบ่งชี้ระดับความเสี่ยงและความเคลื่อนไหวในตลาดในอนาคต

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **VIX ช่วยในการวัดความเปลี่ยนแปลงของตลาดหุ้น โดยเฉพาะในระยะเวลาที่มีความผันผวนมาก ซึ่งสามารถใช้ในการประมาณการความเสี่ยงและความเปลี่ยนแปลงในตลาดที่จะเกิดขึ้นในอนาคต**

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **VIX เป็นดัชนีที่มีค่าระหว่าง 0 ถึง 100 โดยค่าที่ต่ำกว่า 20 ถือว่าตลาดนิ่งและคล่องตัว ในขณะที่ค่าที่สูงกว่า 20 แสดงถึงความเปลี่ยนแปลงและความผันผวนในตลาดที่เพิ่มขึ้น**

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - **VIX มักถูกใช้เป็นตัวบ่งชี้ของความเสี่ยงในตลาดทางการเงิน และใช้ในการวิเคราะห์และตัดสินใจในการลงทุน เช่น การปรับสัดส่วนการลงทุน การจัดการความเสี่ยง หรือการวางแผนการลงทุนในอนาคต**

- ### Vanguard Short-Term Inflation-Protected Securities Index Fund
#### &nbsp;&nbsp;Vanguard Short-Term Inflation-Protected Securities Index Fund (VITP) เป็นกองทุนรวมที่จัดการโดย Vanguard ซึ่งเป็นบริษัทลงทุนที่มีชื่อเสียง กองทุนนี้เน้นการลงทุนในตราสารหนี้ที่ป้องกันการเสื่อมค่าจากการเงินและเสื่อมค่าจากการเพิ่มขึ้นของราคาในอนาคต กลไกที่ใช้ในการป้องกันการเสื่อมค่าคือการลงทุนในพันธบัตรค่าเงินต้นที่เชื่อมโยงกับอัตราการเสื่อมค่าของราคาในตลาด &nbsp;&nbsp; VITP มุ่งเน้นการลงทุนในตราสารหนี้ที่มีระยะเวลาสั้น ๆ โดยทั่วไปเป็นพันธบัตรค่าเงินต้นที่มีกำหนดคืนที่เปลี่ยนแปลงตามอัตราการเสื่อมค่าของราคา (inflation) ดังนั้น การลงทุนใน VITP อาจช่วยให้นักลงทุนปกป้องตนเองจากการเสื่อมค่าในการเงินและความเพิ่มขึ้นของราคาในระยะสั้น ๆ &nbsp;&nbsp; การลงทุนใน VITP สามารถช่วยให้นักลงทุนหากำไรจากดอกเบี้ยตราสารหนี้ที่ปรับตามอัตราการเสื่อมค่าของราคา นอกจากนี้ ค่าธรรมเนียมจัดการของ VITP มีแนวโน้มจะต่ำกว่ากองทุนที่มีการจัดการอื่น ๆ ซึ่งช่วยเพิ่มผลตอบแทนในการลงทุน &nbsp;&nbsp; อย่างไรก็ตาม การลงทุนในกองทุนรวมเสมอมีความเสี่ยง ผลตอบแทนของ VITP สามารถขึ้นอยู่กับสถานการณ์ตลาดและอัตราเปลี่ยนแปลงของราคา การอ่านและศึกษาข้อมูลเกี่ยวกับกองทุนก่อนการลงทุนจึงเป็นสิ่งสำคัญเพื่อให้คุณทำการตัดสินใจที่มีความรับผิดชอบ

- ### U.S. Treasury Bond Futures
#### &nbsp;&nbsp; U.S. Treasury Bond Futures คือสัญญาซื้อขายล่วงหน้าสำหรับพันธบัตรค่าเงินต้นของรัฐบาลสหรัฐอเมริกา (U.S. Treasury Bonds) ในตลาดฟิวเจอร์ส การซื้อขายฟิวเจอร์สเป็นการซื้อขายในสัญญาซึ่งกำหนดให้ซื้อหรือขายพันธบัตรค่าเงินต้นในวันที่กำหนดล่วงหน้า โดยมีราคาและเงื่อนไขการซื้อขายที่ระบุไว้ในสัญญา &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; U.S. Treasury Bond Futures ถูกใช้เป็นเครื่องมือในการป้องกันความเสี่ยงและการซื้อขายของนักลงทุน โดยเฉพาะอย่างยิ่งในการบริหารจัดการความเสี่ยงในอัตราดอกเบี้ย นักลงทุนที่มีผลกระทบต่อการเปลี่ยนแปลงของอัตราดอกเบี้ยสามารถใช้ U.S. Treasury Bond Futures เพื่อป้องกันความเสี่ยงจากการเพิ่มขึ้นหรือลดลงของอัตราดอกเบี้ย &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; การซื้อขาย U.S. Treasury Bond Futures ในตลาดฟิวเจอร์สมักเกิดขึ้นในการซื้อขายหลักทรัพย์ทางการเงินอื่น ๆ ซึ่งสามารถทำได้ผ่านบริษัทหลักทรัพย์ที่มีความเชี่ยวชาญในการซื้อขายฟิวเจอร์ส การซื้อขายในตลาดฟิวเจอร์สอาจมีความซับซ้อนและมีความเสี่ยง นักลงทุนควรทำความเข้าใจเงื่อนไขและความเสี่ยงที่เกี่ยวข้องกับการลงทุนในฟิวเจอร์สก่อนการเข้าร่วมซื้อขาย

## **Independent Variable**
- ### **Open** &nbsp;&nbsp; คือราคาในการเริ่มต้นการซื้อขายหุ้นในแต่ละวัน หรือเป็นราคาที่สัญญาณหุ้นถูกทำการซื้อหรือขายเป็นครั้งแรกในแต่ละวัน การรู้ค่า "Open" ช่วยให้นักลงทุนทราบถึงแนวโน้มเบื้องต้นของการขยับของตลาดในแต่ละวัน

- ### **Volume** &nbsp;&nbsp; รายงานถึงปริมาณการซื้อขายของหุ้นในระยะเวลาที่กำหนด ปริมาณการซื้อขายแสดงถึงความสนใจของตลาดและแรงจูงใจในการซื้อหรือขายหุ้น การวิเคราะห์ปริมาณการซื้อขายสามารถช่วยนักลงทุนตีความความเป็นไปได้ของการขยับของราคา
- ### **High** &nbsp;&nbsp; เป็นราคาสูงสุดที่หุ้นถูกซื้อหรือขายในระยะเวลาที่กำหนด, เช่น ในหนึ่งวัน การรู้ค่า "High" ช่วยให้นักลงทุนเข้าใจถึงแนวโน้มการขยับของราคาหุ้นในแต่ละวันที่สูงที่สุด
- ### **Low** &nbsp;&nbsp; เป็นราคาต่ำสุดที่หุ้นถูกซื้อหรือขายในระยะเวลาที่กำหนด การรู้ค่า "Low" ช่วยให้นักลงทุนทราบถึงระดับราคาที่ตลาดเห็นว่าเป็นจุดที่น่าสนใจในการซื้อในแต่ละวัน

#### S&P 500 Index with other 7 Index

```python
# Specified tickers
index_tickers = ['^TNX', '^VIX', 'CL=F', 'GC=F', '^GSPC', '^TYX', 'VCSH']

# The names for each ticker for better clarity
index_names = {
    '^TNX': '10 Year Bond',
    '^VIX': 'CBOE_Volatility',
    'CL=F': 'Crude_Oil',
    'GC=F': 'Gold',
    '^GSPC': 'S&P',
    '^TYX': 'U.S._Treasury_Bond',
    'VCSH': 'Vanguard_short_term'
}

# Fetch data
data = {}
for ticker in index_tickers:
    ticker_data = yf.download(ticker, start="2021-05-01", end="2023-05-01")

    # Keeping only the desired columns
    ticker_data = ticker_data[['High', 'Low', 'Volume', 'Adj Close']]

    # Renaming 'Adj Close' to 'AdjClose'
    ticker_data = ticker_data.rename(columns={'Adj Close': 'AdjClose'})

    # Append the ticker's name (from the index_names dictionary) as a prefix to each column name for clarity
    ticker_data.columns = [f"{index_names[ticker]}_{col}" for col in ticker_data.columns]

    data[ticker] = ticker_data

# Combining all DataFrames into one
data = pd.concat(data.values(), axis=1)

# Dropping unwanted volume columns for '^TNX' and '^VIX'
data = data.drop(columns=['10 Year Bond_Volume', 'CBOE_Volatility_Volume','U.S._Treasury_Bond_Volume'])

# สร้างกราฟแท่งเทียนด้วย Plotly
fig = go.Figure(data=[go.Candlestick(x=data.index,
                open=data['S&P_AdjClose'],  # Use AdjClose as Open for the sake of visualization
                high=data['S&P_High'],
                low=data['S&P_Low'],
                close=data['S&P_AdjClose'],
                name='Training Data')])

# ปรับขนาดและรูปร่างของกราฟ
fig.update_layout(
    autosize=False,
    width=1900,  # ลดความกว้าง
    height=900,  # เพิ่มความสูง
)

fig.show()
```

### Upon analyzing the dataset using all 7 indices, it was observed that the R2 Score was lower than when analyzing solely with the S&P 500 index. Hence, the model was further subjected to feature importance techniques to scrutinize and select significant coefficients of independent variables.
<br>
<br>

```python
np.round(lr.coef_,3)
```

```python
lr.feature_names_in_
```

```python
import plotly.graph_objects as go

feature_weights = pd.DataFrame({
    'Weight': np.round(lr.coef_.squeeze(), 3)
}, index=lr.feature_names_in_)

sorted_weights = feature_weights.sort_values(by='Weight', ascending=False)

# Conditionally color bars based on the Weight value
colors = ['#FF761D' if -4 <= val <= 4 else '#695BFD' for val in sorted_weights['Weight']]

# Plotting using plotly.graph_objects
fig = go.Figure(go.Bar(
    x=sorted_weights['Weight'],
    y=sorted_weights.index,
    orientation='h',
    marker_color=colors
))

fig.update_layout(title="Feature Weights", autosize=False, width=1700, height=900)

fig.show()
```

```python
selected_columns = sorted_weights[~sorted_weights['Weight'].between(-4, 4)].index.tolist()
selected_columns.append('S&P_AdjClose')
selected_columns.append('S&P_Low')
selected_columns.append('S&P_High')
selected_columns
```
